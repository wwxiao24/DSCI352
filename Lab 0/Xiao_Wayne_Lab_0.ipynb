```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
```

```python
# Create a dataframe with columns named 'x' and 'y' filled with random numbers
df = pd.DataFrame({"x": np.random.normal(size=50), "y": np.random.normal(size=50)})

# Create an ordinary least squares regression on the data and show summary statistics.
# This formula regresses column 'y' on 'x'
ro = sm.OLS.from_formula("y ~ x", df)
res = ro.fit()
print(res.summary())

# Plot the data
sns.regplot('x', 'y', df)
plt.show()
```

```python
DATA_ROOT = 'ALL CSV FILES - 2nd Edition/'

advertising = pd.read_csv(DATA_ROOT+'Advertising.csv')
advertising.info()
advertising.describe()
```

```python
#TV
tv_df = pd.DataFrame({"x": advertising['TV'], "y": advertising['sales']})

# Create an ordinary least squares regression on the data and show summary statistics.
# This formula regresses column 'y' on 'x'
ro = sm.OLS.from_formula("y ~ x", tv_df)
res = ro.fit()
print(res.summary())

# Plot the data
sns.regplot('x', 'y', tv_df)
plt.show()
```

```python
#Radio
radio_df = pd.DataFrame({"x": advertising['radio'], "y": advertising['sales']})

# Create an ordinary least squares regression on the data and show summary statistics.
# This formula regresses column 'y' on 'x'
ro = sm.OLS.from_formula("y ~ x", radio_df)
res = ro.fit()
print(res.summary())

# Plot the data
sns.regplot('x', 'y', radio_df)
plt.show()
```

```python
#Newspaper
news_df = pd.DataFrame({"x": advertising['newspaper'], "y": advertising['sales']})

# Create an ordinary least squares regression on the data and show summary statistics.
# This formula regresses column 'y' on 'x'
ro = sm.OLS.from_formula("y ~ x", news_df)
res = ro.fit()
print(res.summary())

# Plot the data
sns.regplot('x', 'y', news_df)
plt.show()
```

```python
# Compare the p-values, r-square and confidence intervals for the gradient for each of these.
# P-values: All P values are <.001, which would indicate that TV, Radio and Newspaper all have a significant relationship with sales.
# R-square: 0.612, 0.332, 0.052 for TV, Radio, Newspaper respectively. This indicates that TV data explains or correlates to sales to a higher degree than Radio, which correlates more to sales than Newspaper.
# Confidence intervals: [0.042, 0.053], [0.162, 0.243], [0.022, 0.087] for TV, Radio, Newspaper respectively.
# Confidence intervals (cont): .011, .081, .065 are the differences for the confidence intervals, respectively. The smaller the difference, the stronger the trend in the regression. Thus, this corroborates the p and r-square values.
```
